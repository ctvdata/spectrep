```python
exp1 = re.compile('><')
exp2 = re.compile('<')
exp3 = re.compile('>')

s = exp1.sub("> <", s)
s = exp2.sub(" <", s)
s = exp3.sub("> ", s)
s
```

```python
# tokenizer = RegexpTokenizer('\w+|\$[\d\.]+|\S+')
tokenizer = RegexpTokenizer('\w+|<\w*?>|\S+')
tokens = tokenizer.tokenize(s)
tokens
```

```python
tokens = list(set(tokenizer.tokenize(s)))
tokens.sort()
tokens
```
